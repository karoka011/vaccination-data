a.Extraer la información del archivo

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/karoka011/vaccination-data/refs/heads/main/country_vaccinations.csv'
vaccines_df = pd.read_csv(url)
print(vaccines_df.head())
#https://drive.google.com/file/d/1W4pGJnqwK6fpqv9WUlQL722BEe-ha2MO/view
```

b. Estructura y tipos de datos de cada columna

```python
vaccines_df.info()
vaccines_df['date'] = pd.to_datetime(vaccines_df['date'])
vaccines_df.dtypes
print(vaccines_df.head())
```

## Entrega Reto The Learning Gate -Tec de Monterrey

c.Determinar la cantidad de vacunas aplicadas de cada compañia

```python
vaccines2_df = vaccines_df.copy()
vaccines2_df['vaccines'] = vaccines2_df['vaccines'].str.split(', ')
vaccines2_df = vaccines2_df.explode('vaccines')
cantidad_vacunas = vaccines2_df.groupby('vaccines')['total_vaccinations'].sum()
print(cantidad_vacunas)
```

d.Obtener la cantidad de vacunas aplicadas en el mundo


```python
totalVacunas = vaccines_df['total_vaccinations'].sum()
print("El numero total de vacunas aplicadas en el mundo: ",totalVacunas)
```

e.Calcular el promedio de vacunas aplicadas por país

```python
promedioVacunacionPaises =vaccines_df.groupby('country')['daily_vaccinations'].mean()
print(promedioVacunacionPaises)
```

f.Determinar la cantidad de vacunas aplicadas el dia 29/01/2021 en todo el mundo


```python
fechaDeterminada = vaccines_df[vaccines_df['date'] == '2021-01-29']
cantidadVacunacionDia = fechaDeterminada['daily_vaccinations'].sum()
print("Numero de vacunas aplicadas el 29/01/2021: ",cantidadVacunacionDia)
```

g.Crear un nuevo dataframe que contenga los datos originales y una columna derivada con las diferencias de aplicacion entre las columnas daily_vaccionations y daily_vaccinations_raw

```python
conDiferencias = vaccines_df.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
print(conDiferencias.head())
```

h.Obtener el periodo de tiempo entre el registro con fecha mas reciente y el registro más antiguo

```python
periodoTiempo = vaccines_df['date'].max() - vaccines_df['date'].min()
print("Periodo de tiempo entre registros: ",periodoTiempo)
```

i.Crear un datafram que contenga los datos originales y una columna derivada con la cantidad de vacunas utilizadas cada dia

```python
conCantidad = vaccines_df.copy()
conCantidad['canVac'] = conCantidad['vaccines'].str.split(', ').str.len()
print(conCantidad[['vaccines','canVac']].sample(n=10))
print(conCantidad.info())
```

j.Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020

```python
antes20 = vaccines_df.copy()
antes20 = antes20[antes20['date'] < "20/12/2020"]
print(antes20['date'].sample(n=10))
```

K.Obtener un dataframe denominado Pfizer con todos los registros donde se haya utilizado la vacuna Pfizer

```python
Pfizer = vaccines_df.copy()
Pfizer = Pfizer[Pfizer['vaccines'].str.contains('Pfizer')]
print(Pfizer['vaccines'].sample(n=5))
```

i.Almacenar los dataframes generados en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente

```python
with pd.ExcelWriter('resultadosReto.xlsx', engine='openpyxl') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias')
    conCantidad.to_excel(writer, sheet_name='conCantidad')
    antes20.to_excel(writer, sheet_name='antes20')
    Pfizer.to_excel(writer, sheet_name='Pfizer')
```
